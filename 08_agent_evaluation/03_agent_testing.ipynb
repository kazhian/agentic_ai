# Agent Testing and Validation

In this notebook, we demonstrate how to write simple tests for agent behavior, compare outputs against expected results, and capture failure cases.

```python
from utils.evaluation_tools import compare_responses, success_rate

test_cases = [
    {
        'prompt': 'Summarize AI agent evaluation in one sentence.',
        'expected': 'Describe how to assess agent behavior using tracing and metrics.',
        'actual': 'Explain the process of evaluating an AI agent using logs and metrics.'
    },
    {
        'prompt': 'List three benefits of performance evaluation.',
        'expected': 'Detect regressions, improve reliability, and surface failure modes.',
        'actual': 'Find bugs, reduce cost, and make agents safer.'
    },
]
passes = []
for case in test_cases:
    similarity = compare_responses(case['expected'], case['actual'])['overlap']
    passes.append(similarity > 0.4)
    print(f"Prompt: {case['prompt']}")
    print(f"Similarity: {similarity:.2f}")
    print('---')

print('Suite success rate:', success_rate(passes))
```

## Testing patterns

- Define expected outputs for representative prompts.
- Compare actual agent responses using similarity or semantic metrics.
- Track pass/fail results across versions or agent configurations.
