# Evaluation Dashboard and Reporting

This notebook shows how to assemble evaluation data into a dashboard-style report for agent performance.

```python
from utils.evaluation_tools import success_rate, average_score

evaluation_data = [
    {'run': 'baseline', 'success': True, 'score': 0.82, 'tool_calls': 3},
    {'run': 'version-2', 'success': False, 'score': 0.65, 'tool_calls': 5},
    {'run': 'version-3', 'success': True, 'score': 0.94, 'tool_calls': 2},
]

all_scores = [row['score'] for row in evaluation_data]
all_success = [row['success'] for row in evaluation_data]

print('Average score:', average_score(all_scores))
print('Overall success rate:', success_rate(all_success))

for row in evaluation_data:
    print(f"Run {row['run']}: score={row['score']}, tool_calls={row['tool_calls']}, passed={row['success']}")
```
