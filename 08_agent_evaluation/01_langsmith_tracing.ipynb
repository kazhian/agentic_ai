# LangSmith Tracing for Agent Evaluation

This notebook introduces LangSmith tracing for LangChain agents. You will learn how to capture execution traces, review tool usage, and inspect agent decisions.

```python
from langchain_openai import ChatOpenAI
from langchain.agents.factory import create_agent
from langchain_core.messages import HumanMessage
from utils.langsmith_helpers import create_langsmith_tracer

# Simple echo tool implementation

def echo(text: str) -> str:
    return f"ECHO: {text}"

# Configure a tracer and attach it to the model
tracer = create_langsmith_tracer(project_name="agentic-ai-evaluation")
llm = ChatOpenAI(model="gpt-4o", temperature=0, callbacks=[tracer])
tools = [
    {
        "name": "echo",
        "description": "Simple echo tool for demonstration.",
        "func": echo,
    }
]
agent = create_agent(
    llm,
    tools=tools,
    system_prompt="You are an agent evaluation assistant.",
    debug=True,
)
response = agent.invoke(
    {
        "messages": [
            HumanMessage(
                content="Explain the phrase 'agent evaluation' and use the tool if helpful."
            )
        ]
    }
)
print(response)
```
