# Evaluation Metrics for Agent Performance

This notebook defines common evaluation metrics for agent workflows and shows how to compute them for example outputs.

```python
from utils.evaluation_tools import success_rate, average_score, compare_responses

# Example results from an evaluation suite
results = [True, False, True, True, False]
scores = [0.8, 0.5, 0.9, 0.95, 0.6]

print("Success rate:", success_rate(results))
print("Average score:", average_score(scores))

reference = "The agent should summarize the document in three sentences."
candidate = "Summarize the content in three concise sentences."
print(compare_responses(reference, candidate))
```

## Metric categories

- **Task success**: Did the agent accomplish the required task?
- **Response quality**: How well does the output match the expected answer?
- **Tool efficiency**: How many tool calls were used, and were they necessary?
- **Latency and cost**: How fast and expensive was the execution?
